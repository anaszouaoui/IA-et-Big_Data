# 2.   traitement des valeurs manquantes

```python
na
```

```python
# Calculer la moyenne de 'Sleep Duration' par groupe 'Quality of Sleep'
average_sleep_duration = na.groupby('Quality of Sleep')['Sleep Duration'].mean()

# Remplacer les valeurs manquantes dans 'Sleep Duration' par la moyenne correspondante
na['Sleep Duration'] = na.apply(
    lambda row: average_sleep_duration[row['Quality of Sleep']] if pd.isnull(row['Sleep Duration']) else row['Sleep Duration'],
    axis=1
)

# Afficher le DataFrame mis à jour
print(na)
```

```python
na.isnull().sum()
```

# 3. détectation des valeurs aberrante

```python
column_to_process = 'Heart Rate'
# Calculer le quartile 1 (Q1) et le quartile 3 (Q3)
Q1 = na[column_to_process].quantile(0.25)
Q3 = na[column_to_process].quantile(0.75)
# Calculer l'IQR (plage interquartile)
IQR = Q3 - Q1
# Définir les seuils pour détecter les valeurs aberrantes
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Sélectionner les valeurs aberrantes
outliers = na[(na[column_to_process] < lower_bound) | (na[column_to_process] > upper_bound)]
# Afficher les valeurs aberrantes
print("Valeurs aberrantes dans la colonne '{}':".format(column_to_process))
print(outliers)
```

```python
sns.boxplot(x=na['Heart Rate'])
```

```python
# Convertir la colonne 'sleep duration' en nombres (remplacer les non-numériques par NaN)
na['Quality of Sleep'] = pd.to_numeric(na['Heart Rate'], errors='coerce')

# Filtrer les lignes avec des valeurs manquantes dans la colonne "sleep duration"
lignes_manquantes = na[na['Quality of Sleep'].isnull()]

# Afficher les lignes résultantes
print(lignes_manquantes)
```

# 2. Prétraitement des données

```python
sns.heatmap(na.isna(),cmap='BuPu')
```

```python
na.columns
```

```python
na['Blood Pressure'].unique()
```

```python
na['Blood Pressure']=na['Blood Pressure'].apply(lambda x:0 if x in ['120/80','126/83','125/80','128/84','129/84','117/76','118/76','115/75','125/82','122/80'] else 1)
```

```python
na["Age"]=pd.cut(na["Age"],2)
na["Heart Rate"]=pd.cut(na["Heart Rate"],4)
na["Daily Steps"]=pd.cut(na["Daily Steps"],4)
na["Sleep Duration"]=pd.cut(na["Sleep Duration"],3)
na["Physical Activity Level"]=pd.cut(na["Physical Activity Level"],4)
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

categories=['Gender','Age','Occupation','Sleep Duration','Physical Activity Level','BMI Category','Heart Rate','Daily Steps','Sleep Disorder']
for label in categories:
    na[label]=LE.fit_transform(na[label])
```

```python
na.drop(['Person ID'], axis=1, inplace=True)
```

```python
correlation=na.corr()
max_6_corr=correlation.nlargest(6,"Sleep Disorder")
sns.heatmap(max_6_corr,annot=True,fmt=".2F",annot_kws={"size":8},linewidths=0.5,cmap='BuPu')
plt.title('Un maximum de six caractéristiques affectent le trouble du sommeil')
plt.show()
```

# 2.   Fractionnement des données

```python
x=na.iloc[:,:-1]
y=na.iloc[:,-1]

x_shape=colored(x.shape, "magenta",None, attrs=["blink"])
y_shape=colored(y.shape, "magenta",None, attrs=["blink"])
print('The dimensions of x is : ',x_shape)
print('The dimensions of y is : ',y_shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=32,shuffle=True)
```

```python
x_train_shape=colored(x_train.shape, "magenta",None, attrs=["blink"])
x_test_shape=colored(x_test.shape, "magenta",None, attrs=["blink"])
y_train_shape=colored(y_train.shape, "magenta",None, attrs=["blink"])
y_test_shape=colored(y_test.shape, "magenta",None, attrs=["blink"])

print("x train dimensions :",x_train_shape)
print("x test dimensions: ",x_test_shape)
print("y train dimensions :",y_train_shape)
print("y test dimensions :",y_test_shape)
```

# V) Modélisation

# 1. Modèle de régression logistique

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression().fit(x_train,y_train)
```

```python
LR_training_score=colored(round(LR.score(x_train,y_train)*100,2), "magenta",None, attrs=["blink"])
LR_testing_score=colored(round(LR.score(x_test,y_test)*100,2), "magenta",None, attrs=["blink"])

print(f"LR training score :",LR_training_score)
print("LR testing score :",LR_testing_score)
```




# 2.   Modèle XGBClassifier



```python
from xgboost import XGBClassifier
xgb=XGBClassifier().fit(x_train,y_train)
```

```python
xgb_training_score=colored(round(xgb.score(x_train,y_train)*100,2), "magenta",None, attrs=["blink"])
xgb_testing_score=colored(round(xgb.score(x_test,y_test)*100,2), "magenta",None, attrs=["blink"])

print("xgb training score :",xgb_training_score)
```
